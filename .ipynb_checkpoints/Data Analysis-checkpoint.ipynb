```python
import pandas as pd #Import pandas
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_cleaned_2021.csv')#Load the DataSet to the Notebook
df.info()#Take a wick look at missing values
```

```python
#We create a list of values that we want in Boolean format

bool_columns = ['Hourly', 'Employer provided','Python','spark','aws','excel',\
                'sql','sas','keras','pytorch','scikit','tensor',\
               'hadoop','tableau','bi','flink','mongo','google_an',]
#Loop in all the boolean columns
for i in bool_columns:
    df[i] = df[i].astype('bool')
    
```

```python
#Tasnform Founded column to date format
df['Founded'] = pd.to_datetime(df['Founded'])
```

```python
df.head()
```

```python
#We taeke a look at the column of job_title_sim
df['job_title_sim'].value_counts()
```

```python
# Drop the colums with na in job title simplify
df =df[df['job_title_sim']!='na']
df.shape
```

```python
#Take a look at the Degree column
df['Degree'].value_counts()
```

```python
#Replace na values with not specify
df['Degree'] =df['Degree'].replace('na','not specify')
df.head(3)
```

```python
#We take a look at the sector column
df['Sector'].value_counts()
```

```python
#we drop the missing values
df = df[df['Sector']!='-1']
df.shape
```

```python
#We take a look at Job Location column
df['Job Location'].value_counts()
```

```python
#We take a look at duplicated values
df[df.duplicated()]
```

```python
#Take a look at the average salary outliers
plt.boxplot(df['Avg Salary(K)'])
```

```python
#Takeing a closer look at the outliers
df[df['Avg Salary(K)']>200]
```

```python
#Save the clean Data Set as a csv
df.to_csv('DS_Jobs.csv')
```
